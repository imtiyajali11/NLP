```python
a = tf.constant(10)      #scalar
print ("a :",type(a))
print(a.shape)
print(a)

b = tf.constant([1,2,3])    #scalar
print ("b :",type(b))
print(b.shape)
print(b)

c = tf.constant([[1,2,3],[4,5,6]]) # matrix
print("c :", type(c))
print(c.shape)
print(c)
```

```python
## Operations

a= tf.constant(1)
b = tf.constant(2)
c = tf.add(a,b)
print(c)

d=tf.multiply(a,b)
print(d)

e = tf.subtract(a,b)
print(e)

f = tf.divide(a,b)
print(f)
```

```python
## Conversion numpy to tensor
import numpy as np

array = np.array([[1,2,3],[4,5,6]])

print(array)

tensor = tf.convert_to_tensor (array, dtype = tf.float32)
print(type(tensor))
```

```python
# ANN

x = np.array([1,2,3,4,5],dtype=float )
y = np.array([6,7,8,9,10],dtype=float )

model = tf.keras.Sequential()     #Class initiate
model.add(tf.keras.layers.Dense(units =100, input_shape = [1],activation='relu'))    # add layers
model.add(tf.keras.layers.Dense(units =10, activation='relu'))                      # hidden layers
model.add(tf.keras.layers.Dense(units =1,activation='linear'))                       #output layers


model.compile(optimizer = 'adam', loss = 'mean_squared_error')       # configure
```

```python
model.fit(x,y,epochs = 10)
```

```python
a = np.array ([6.0], dtype = float )
```

```python
model.predict(a)
```

```python
print(model.summary())
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(model.history.history['loss'], label = 'Train loss')
plt.legend()
plt.show()
```
